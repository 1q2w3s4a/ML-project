After get the small size data for train and test, we follow the example in MNIST neural network demo, clear the keras session. Then, create a neural network with 256 hidden layer and sigmoid activation.

```python

from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Activation
import tensorflow.keras.backend as K
K.clear_session()

nout = np.max(ytr)+1
nh = 256
model = Sequential()
model.add(Dense(nh, input_shape=(1,), activation='sigmoid', name='hidden'))
model.add(Dense(nout, activation='softmax', name='output'))
model.summary()
```

We create the optimizer and compile the modle. We select the learning rate as 0.001 and coose mean_square_error as loss function

```python

from tensorflow.keras import optimizers
K.clear_session()

nout = np.max(ytr)+1
nh = 256
model1 = Sequential()
model1.add(Dense(nh, input_shape=(1,), activation='sigmoid', name='hidden'))
model1.add(Dense(nout, activation='softmax', name='output'))
model1.summary()
opt = optimizers.Adam(lr=0.001)
model1.compile(optimizer=opt,
              loss='mean_squared_error',metrics=['accuracy'])
hist = model1.fit(Xtr,ytr,epochs=10, batch_size = 100, validation_data = (Xts,yts))
```

```python
val_acc = hist.history['val_acc']
nepochs = len(val_acc)
plt.plot(np.arange(1,nepochs+1), val_acc, 'o-', linewidth=2)
plt.grid()
plt.xlabel('Epoch')
plt.ylabel('Test accuracy')
```

We plot the validation accuracy saved in hist.history dictionary. It gives one accuracy value per epoch.

```python
loss = hist.history['loss']
plt.semilogy(np.arange(1,nepochs+1), loss, 'o-')
plt.xlabel('Epochs', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.grid()
plt.tight_layout()
```

The accuracy is shift around 0.33 and we can see the predicting error is very large. One possible reason is the too precise of time_to_failure increased the predicting error. Threfore we try to only look the three decimal digits to see the error rate.

```python
from decimal import *
ytr1 = [float(Decimal("%.3f" % e)) for e in ytr]
yts1 = [float(Decimal("%.3f" % e)) for e in yts]
```
